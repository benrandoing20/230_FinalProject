## Create 2 separate folders to keep the sorted train and test datasets

```python
def sort_data(df_metadata, base_path):

    # Check if path exists
    for index, row in train_metadata.iterrows():
        sorted_data_path = os.path.join(base_path, "sorted_data", "train", row["benign_malignant"])
        print(sorted_data_path)
        if not os.path.exists(sorted_data_path):
            os.popen(f"mkdir {sorted_data_path}")

        curr_image_path = os.path.join(base_path, "train", row["image_name"]+".jpg")
        new_image_path = os.path.join(sorted_data_path, row["image_name"]+".jpg")

        if os.path.exists(curr_image_path):
            shutil.copy2(curr_image_path, new_image_path)
```
